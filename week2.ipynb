```python
import cv2 as cv
import matplotlib.pyplot as plt

# Membaca gambar dalam warna asli
img = cv.imread('/content/drive/MyDrive/images.jpeg')

# Membuat citra baru hanya dengan saluran Red-Blue
red_blue = img.copy()
red_blue[:, :, 1] = 0  # Set saluran hijau menjadi 0 (hanya tinggalkan saluran merah dan biru)

# Menampilkan citra dalam saluran Red-Blue
plt.imshow(cv.cvtColor(red_blue, cv.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Red-Blue Channel')
plt.show()
```

```python
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread('/content/drive/MyDrive/images.jpeg')

# Tentukan rentang baris dan kolom yang Anda inginkan
baris_mulai = 20
baris_akhir = 115
kolom_mulai = 25
kolom_akhir = 120

# Potong gambar sesuai dengan rentang yang Anda inginkan
gambar_potong = img[baris_mulai:baris_akhir, kolom_mulai:kolom_akhir]

# Tampilkan gambar yang telah dipotong
plt.imshow(cv.cvtColor(gambar_potong, cv.COLOR_BGR2RGB))  # Konversi warna dari BGR ke RGB untuk tampilan matplotlib
plt.axis('off')  # Untuk menghilangkan sumbu x dan y
plt.show()
```

```python
import cv2 as cv
import matplotlib.pyplot as plt

# Baca gambar menggunakan OpenCV
img = cv.imread('/content/drive/MyDrive/images.jpeg')

# Tentukan rentang baris yang Anda inginkan (5-30) dan semua kolom
baris_mulai = 5
baris_akhir = 30

# Potong gambar sesuai dengan rentang baris yang Anda inginkan
gambar_potong = img[baris_mulai:baris_akhir, :]

# Pisahkan saluran merah
saluran_merah = gambar_potong.copy()
saluran_merah[:, :, 1] = 0  # Setel saluran hijau (G) dan biru (B) menjadi 0

# Tampilkan saluran merah yang telah dipotong
plt.imshow(cv.cvtColor(saluran_merah, cv.COLOR_BGR2RGB))  # Konversi warna dari BGR ke RGB untuk tampilan matplotlib
plt.axis('off')  # Untuk menghilangkan sumbu x dan y
plt.show()
```
